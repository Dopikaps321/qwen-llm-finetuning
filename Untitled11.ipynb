```python
# CELL 2: Imports
import json
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    BitsAndBytesConfig,
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer

print("✓ All imports successful!")
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
```

```python
# CELL 3: Upload Dataset
from google.colab import files

print(" Please upload your JSONL file:")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
print(f"\n✓ File uploaded: {filename}")

# Load JSONL data
data = []
with open(filename, 'r') as f:
    for line in f:
        data.append(json.loads(line))

print(f"✓ Loaded {len(data)} samples")

# Display sample to verify structure
print("\n Sample data structure:")
print(json.dumps(data[0], indent=2)[:800] + "...")
```

```python
# CELL 4: Verify Dataset Structure
sample = data[0]

print(" Verifying dataset structure...")
print(f"Has 'input' key: {'input' in sample}")
print(f"Has 'output' key: {'output' in sample}")

if 'input' in sample and 'output' in sample:
    print(" Dataset structure is CORRECT!")
    print(f"\nInput keys: {list(sample['input'].keys())}")
    print(f"Output keys: {list(sample['output'].keys())}")
else:
    print(" ERROR: Dataset structure is INCORRECT!")
    print("Expected structure: {'input': {...}, 'output': {'summary': '...'}}")
    print(f"Actual keys: {list(sample.keys())}")
    raise ValueError("Dataset structure mismatch!")
```

```python
# CELL 5: Format Dataset for Training (CONTROLLED ROOT CAUSE)

def format_root_cause_instruction(input_data, root_cause):
    """
    Format claim data into instruction-following format for
    ROOT CAUSE generation (controlled reasoning)
    """
    claim_details = json.dumps(input_data, indent=2)

    formatted_text = f"""<|im_start|>system
You are a medical claims specialist.

Generate the ROOT CAUSE of denial.

Rules:
- Focus ONLY on denied CPT lines.
- Explain WHY the denial occurred.
- Base the explanation strictly on denial codes and denial reason text.
- Do NOT summarize the entire claim.
- Do NOT mention approved CPT lines unless necessary for context.
- Do NOT recommend actions or next steps.
- Do NOT introduce causes not explicitly supported by the input.
- If the available information is insufficient, state that the root cause cannot be clearly determined.
<|im_end|>
<|im_start|>user
Analyze the following medical claim and provide the ROOT CAUSE for denial:

{claim_details}<|im_end|>
<|im_start|>assistant
{root_cause}<|im_end|>"""

    return formatted_text
```

```python
formatted_data = []
skipped = 0

for item in data:
    try:
        input_data = item["input"]
        root_cause = item["output"]["root_cause"]
        formatted_text = format_root_cause_instruction(input_data, root_cause)
        formatted_data.append({"text": formatted_text})
    except KeyError as e:
        print(f"Skipping record due to missing key: {e}")
        skipped += 1

print(f"Formatted {len(formatted_data)} samples ({skipped} skipped)")
```

```python
dataset = Dataset.from_list(formatted_data)
dataset = dataset.train_test_split(test_size=0.1, seed=42)

print("Train size:", len(dataset["train"]))
print("Validation size:", len(dataset["test"]))
```

```python
# CELL 7: Load Tokenizer and Model
model_name = "Qwen/Qwen2.5-1.5B-Instruct"

print(f" Loading model: {model_name}")

# Configure 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    trust_remote_code=True
)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

print("✓ Tokenizer loaded!")

# Load model with quantization
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)

# Prepare model for k-bit training
model = prepare_model_for_kbit_training(model)

print(" Model loaded and prepared for training!")
print(f"Model device: {next(model.parameters()).device}")
```

```python
# CELL 8: Configure LoRA
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj"
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)

print(" LoRA applied!")
print("\n Trainable Parameters:")
model.print_trainable_parameters()
```

```python
# CELL 9: Configure Training Arguments
training_args = TrainingArguments(
    output_dir="./qwen-medical-claims",
    num_train_epochs=3,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    lr_scheduler_type="cosine",
    warmup_ratio=0.1,
    logging_steps=10,
    save_strategy="epoch",
    eval_strategy="epoch",
    bf16=True,
    optim="paged_adamw_8bit",
    save_total_limit=2,
    load_best_model_at_end=True,
    report_to="none",
    gradient_checkpointing=True,
    max_grad_norm=0.3,
    weight_decay=0.001,
)
```

```python
# CELL 10: Tokenize Dataset
def tokenize_function(examples):
    """Tokenize the text field"""
    return tokenizer(
        examples["text"],
        truncation=True,
        padding=False,
        max_length=2048,
    )

print(" Tokenizing dataset...")
tokenized_train = dataset["train"].map(
    tokenize_function,
    batched=True,
    remove_columns=dataset["train"].column_names
)
tokenized_eval = dataset["test"].map(
    tokenize_function,
    batched=True,
    remove_columns=dataset["test"].column_names
)
print(" Tokenization complete!")

# CELL 11: Initialize Trainer and Start Training
from transformers import Trainer, DataCollatorForLanguageModeling

# Data collator for causal language modeling
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # We're doing causal LM, not masked LM
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
    data_collator=data_collator,
)


# Start training
trainer.train()
```

```python
# CELL 12 — Merge LoRA weights
print("Merging LoRA weights...")
merged_model = model.merge_and_unload()
print("Merge complete!")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Mount Drive if not already mounted
from google.colab import drive
drive.mount('/content/drive')

MODEL_PATH = "/content/drive/MyDrive/qwen-2.5B-finetuned"

print("Loading model...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    torch_dtype=torch.float16,
    device_map="auto"
)
model.eval()
print("Model loaded and ready!")
```

```python
import json

def generate_root_cause(input_claim: dict) -> str:
    prompt = (
        "<|im_start|>system\n"
        "You are a medical claims specialist. Identify the ROOT CAUSE of denial.\n"
        "<|im_end|>\n"
        "<|im_start|>user\n"
        "Analyze the following medical claim and provide the ROOT CAUSE for denial:\n\n"
        f"{json.dumps(input_claim, indent=2)}<|im_end|>\n"
        "<|im_start|>assistant\n"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        output_ids = model.generate(
            **inputs,
            max_new_tokens=300,
            do_sample=False,
            repetition_penalty=1.1,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.eos_token_id,
        )

    new_tokens = output_ids[0][inputs["input_ids"].shape[-1]:]
    return tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
```

```python
import gradio as gr

SAMPLE = json.dumps({
    "claim_id": 145446,
    "total_cpt_lines": 2,
    "cpt_lines": [
        {
            "status": "DENIED",
            "cpt_code": "47563",
            "pos": "21",
            "icd_codes": ["K80.10"],
            "denial_reasons": [{"code": "88", "reason": "Overpaid"}],
            "modifiers": []
        }
    ]
}, indent=2)

def run_inference(json_text: str) -> str:
    try:
        claim = json.loads(json_text)
    except Exception:
        return "Invalid JSON — please paste a valid claim input."
    return generate_root_cause(claim)

demo = gr.Interface(
    fn=run_inference,
    inputs=gr.Textbox(lines=20, label="Paste Claim JSON", value=SAMPLE),
    outputs=gr.Textbox(lines=10, label="Root Cause"),
    title="Medical Claim Root Cause Analyzer",
    description="Fine-tuned Qwen 2.5-1.5B — paste your claim JSON and click Submit",
)

demo.launch(share=True)  # share=True gives a public link in Colab
```
